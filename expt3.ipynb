```python
import pandas as pd
import numpy as np

# Load dataset
data = pd.read_csv("DigitalAd_dataset.csv")

# Display the shape and first 5 rows of the dataset
print(data.shape)
print(data.head(5))

# Drop unnecessary columns
data = data.drop(columns=["Unnamed: 11", "Unnamed: 12", "Unnamed: 13", "Unnamed: 14", "Unnamed: 15"])

# Handle categorical columns by encoding them
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

# Encode binary categorical columns using LabelEncoder
label_encoder = LabelEncoder()
data['Gender'] = label_encoder.fit_transform(data['Gender'])

# Apply OneHotEncoding to non-binary categorical columns like 'Device', 'Day', 'Channel'
# Weâ€™ll use a ColumnTransformer to apply OneHotEncoding to the specified columns.
# The rest will remain unchanged (no encoding).

column_transformer = ColumnTransformer(
    transformers=[
        ('device_day_channel', OneHotEncoder(), ['Device', 'Day', 'Channel']),
    ],
    remainder='passthrough'  # Keep other columns as is
)

# Apply the transformation
X = column_transformer.fit_transform(data.iloc[:, :-1])  # All columns except the target column
Y = data.iloc[:, -1].values   # Last column as target

# Split the data into training and testing sets (75% training, 25% testing)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=0)

# Feature scaling (standardizing the features)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Train the Logistic Regression model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=0)
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model using confusion matrix and accuracy
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix: ")
print(cm)
print("Accuracy of the Model: {0}%".format(accuracy_score(y_test, y_pred) * 100))

# Predict for a new customer
age = int(input("Enter New Customer Age: "))
sal = int(input("Enter New Customer Salary: "))
newCust = [[age, sal]]
result = model.predict(sc.transform(newCust))
print(result)

# Output the prediction result
if result == 1:
    print("Customer will Buy")
else:
    print("Customer won't Buy")
```

```python
import pandas as pd
import numpy as np

# Load dataset
data = pd.read_csv("DigitalAd_dataset.csv")

# Display the shape and first 5 rows of the dataset
print("Dataset Shape:", data.shape)
print("First 5 Rows of Data:")
print(data.head(5))

# Check for missing values or non-numeric data
print("Check for missing values or non-numeric data:")
print(data.info())

# Separate features (X) and target (Y)
X = data.iloc[:, :-1]  # Features (all columns except last one)
Y = data.iloc[:, -1]   # Target (last column)

# Convert categorical columns to numeric using one-hot encoding (if necessary)
# Check which columns are of type object (categorical)
categorical_columns = X.select_dtypes(include=['object']).columns

# Apply one-hot encoding for categorical columns (convert to dummy variables)
X = pd.get_dummies(X, columns=categorical_columns, drop_first=True)

# Now we can safely apply scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

# Apply scaling only to numeric columns
X_scaled = sc.fit_transform(X)  # This scales all numeric features

# Split the data into training and testing sets (75% training, 25% testing)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, Y, test_size=0.25, random_state=0)

# Train the Logistic Regression model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=0)
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model using confusion matrix and accuracy
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix: ")
print(cm)
print("Accuracy of the Model: {0}%".format(accuracy_score(y_test, y_pred) * 100))

# Predict for a new customer with validation on the input
while True:
    try:
        age = int(input("Enter New Customer Age: "))
        sal = int(input("Enter New Customer Salary: "))
        if age <= 0 or sal <= 0:
            print("Age and Salary should be positive values. Please try again.")
            continue
        newCust = [[age, sal]]
        
        # Apply scaling to the new customer input (same scaling as for training data)
        newCust_scaled = sc.transform(newCust)
        
        result = model.predict(newCust_scaled)
        print("Prediction Result:", result)

        # Output the prediction result
        if result == 1:
            print("Customer will Buy")
        else:
            print("Customer won't Buy")
        break  # Break the loop once valid inputs are given
    except ValueError:
        print("Invalid input! Please enter valid numbers for Age and Salary.")
```
